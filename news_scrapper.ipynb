**Extract the blog title from each
country page given in the menu of the website**


```python
from bs4 import BeautifulSoup as BS
import requests
import pandas as pd 
import os
```

```python
url= 'https://edition.cnn.com/'
reponse=requests.get(url)
```

```python
mp_soup= BS(reponse.content,'html.parser')
```

```python
filter1=mp_soup.find_all('ul', class_= "subnav__subsections")[0] #get the unorder-list where the country names are stored

# print(len(filter1))
```

```python
filter_list=filter1.find_all('a', href=True) # create a list of attributes
country_links={}                        # creat a hashmap to store the country name and the href link 

for item in filter_list:
    country_links[item.text.strip()] = item['href']
print(country_links)
    



```

**Created a Demo dataframe with one input**

```python
temp_url= country_links['Africa']
temp_response=requests.get(temp_url)
temp_soup=BS(temp_response.content, 'html.parser')
temp_list=[]

for temp_items in temp_soup.find_all('span', class_='container__headline-text'):
    temp_list.append(temp_items.text)
    pd.set_option('display.colheader_justify', 'center')
df=pd.DataFrame(temp_list, columns=['Africa'])

df.index=range(1, len(df)+1)
df.index.name='Sr.no'
display(df.style.set_properties(**{'text-align': 'left'}).set_table_styles([dict(selector='th', props=[('text-align', 'left')])]))

```

```python
for links in country_links.values():
    temp_url= links
    temp_response=requests.get(temp_url)
    temp_soup=BS(temp_response.content, 'html.parser')
    temp_list=[]

    for temp_items in temp_soup.find_all('span', class_='container__headline-text'):
        temp_list.append(temp_items.text)
        pd.set_option('display.colheader_justify', 'center')
    df=pd.DataFrame(temp_list, columns=['News'])

    df.index=range(1, len(df)+1)
    df.index.name='Sr.no'
    display(df.style.set_properties(**{'text-align': 'left'}).set_table_styles([dict(selector='th', props=[('text-align', 'left')])]))
    file_name = str(list(country_links.keys())[list(country_links.values()).index(temp_url)]) + '.csv'
    df.to_csv(file_name, encoding='utf-8-sig')
```

**TO make a master data CSV file**


```python
#Read data from the files in the folder and make a master file of all the data
directory = r'C:\Users\Vedaant\Desktop\Assignment pixelo'


merged_df = pd.DataFrame()


for filename in os.listdir(directory):
    if filename.endswith(".csv"):  
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        second_column = df.iloc[:, 1]
        df_to_concat = pd.DataFrame({'News': second_column})
        df_to_concat['Country'] = filename.strip('.csv')
        merged_df = pd.concat([merged_df, df_to_concat], ignore_index=True)


```

```python
# Save the dataframe to a new CSV file
merged_df.index=range(1, len(merged_df)+1)
merged_df.index.name='Sr.no'
display(merged_df)
filename= "MasterFile.csv"
merged_df.to_csv(filename ,encoding='utf-8-sig')
```
